```python
import matplotlib.pyplot as plt

methods = [
    {
        "method_name": "baseline (std::map)",
        "folder_name": "results/baseline/"
    },
    {
        "method_name": "std::unordered_map",
        "folder_name": "results/std_unordered/"
    },
    {
        "method_name": "hand-written map with Hash = std::hash",
        "folder_name": "results/fast_std_hash/"
    }
]

for method in methods:
    try:
        with open(method["folder_name"] + 'results_times.txt', 'r') as file:
            content = file.read().strip()
            method["result_times"] = [float(time) for time in content.split(',')]
    except Exception as e:
        print(f"An unexpected error occurred: {e}")


file_indices = range(1, len(methods[0]["result_times"]) + 1)

plt.figure(figsize=(10, 6))
for method in methods:
    plt.plot(file_indices, method["result_times"], label=method["method_name"], linestyle='-')
plt.title('Result Time Comparison')
plt.xlabel('File Size in MB')
plt.ylabel('Result Time (Seconds)')
plt.legend()
plt.show()
```

```python
for method in methods:
    try:
        with open(method["folder_name"] + 'results_size.txt', 'r') as file:
            content = file.read().strip()
            method["result_sizes"] = [float(time) for time in content.split(',')]
    except Exception as e:
        print(f"An unexpected error occurred: {e}")


plt.figure(figsize=(10, 6))
for method in methods:
    plt.plot(file_indices, method["result_sizes"], label=method["method_name"], marker='o', linestyle='-')

plt.title('Result Time Comparison')
plt.xlabel('File Index')
plt.ylabel('Result Time')
plt.legend()
plt.show()
```
